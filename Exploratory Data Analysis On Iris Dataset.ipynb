# Iris Data For Exploratory Data Analysis

Here are the some methods that is easy and useful for the EDA.  
Use these method could basically summarize and overview the data before we make the further inference.

Seperate the two parts:  
> **First 1** is the discriptive summary.  
> **Second** is the data visualization to view the data.  

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sc
%pylab inline
```

```python
# import dataset
from sklearn.datasets import load_iris
dataset = load_iris()
```

```python
dataset.target
```

```python
dataset.data
```

```python
dataset.feature_names
```

```python
dataset.target_names
```

```python
iris = pd.DataFrame(dataset.data, columns=['sepal length','sepal width','petal length','petal width'])
```

```python
iris.head()
```

# Basic Descriptive Statistic

```python
#Basis Statistic Summary 
iris.describe()
```

```python
#Basic summary with 7-number summaries
iris.describe(percentiles=[.125, .25, .5, .75, .875])
```

```python
#Show the skewness and kurtosis
skewness = sc.stats.skew(iris)
kurtosis = sc.stats.kurtosis(iris)

print('Skewness:',skewness)
print('kurtosis:',kurtosis)
```

```python
target = pd.Series(dataset.target)
```

```python
#View the counts of the target
target.value_counts()
```

# Data Visualization

## Boxplot 
Visualize the data distribution.  
Get the basic overview of the data.

```python
sns.set(style="whitegrid")
iris.boxplot()
```

```python
#Violinplot, the boxplot mix the distribution
sns.violinplot(iris)
```

## Historgram
View the distribution of the data

```python
#Overlap in the one picture
iris.plot(kind='hist', alpha = 0.7)
```

```python
#Seperate to the four groups
iris.hist()
```

## Scatter Matrix
View the relationship between the two parameter

```python
g = sns.PairGrid(iris)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot, cmap="Blues_d") 
g.map_diag(sns.kdeplot, lw=3, legend=False)
```

## Heatmap
View the correlation between the feature

```python
cmap = sns.light_palette((260, 75, 60),input="husl", as_cmap=True)
sns.heatmap(iris.corr(), annot=True, cmap=cmap)
```

## Parallel Coordinates
Another way to overview the multivariate data.  
Show the all data points and connect together.

```python
iris['Spicies'] = dataset.target

iris['Spicies']=iris['Spicies'].replace({0:'setosa',1:'versicolor',2:'virginica'})

from pandas.tools.plotting import parallel_coordinates
parallel_coordinates(iris,'Spicies')
```
