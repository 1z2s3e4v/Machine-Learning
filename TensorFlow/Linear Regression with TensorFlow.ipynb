# Data Modeling

```python
import numpy as np
```

```python
#Define the random numbers for x and y
number_of_points = 500
x_point = []
y_point = []

a = 0.22
b = 0.78

for i in range(number_of_points):
    x = np.random.uniform(0.0, 0.5)
    y = a*x+b+np.random.uniform(0.0,0.1)
    x_point.append(x)
    y_point.append(y)
```

```python
import matplotlib.pyplot as plt
```

```python
#Visualize the data
plt.plot(x_point, y_point, 'o', label='Input Data')
plt.legend()
plt.show()
```

# Cost functions and gradient descent

Linear Regression Learing Step:  
1. Define the variable of parameter
2. Define the variable of constant
3. Define the function
4. Define the cost function
5. Call the gradient descent function and assign the learning rate
6. Minimize the cost_function
7. Use the session to run the optimizer and assign the training time (use for function)

```python
import tensorflow as tf
```

```python
#Define variable
#Define a as uniform between -1 to 1
A = tf.Variable(tf.random_uniform([1],-1,1))

#Define the constant b as 0
b = tf.Variable(tf.zeros([1]))
```

```python
#Define the linear relationship
y = A*x_point + b
```

```python
#Define cost function
#Evaluate how well the parameters are correct
cost_function = tf.reduce_mean(tf.square(y-y_point))
```

```python
#Use gradient descent algorithm:
optimizer = tf.train.GradientDescentOptimizer(0.5) #learning rate = 0.5

#Define the train as the result of application of cost function
#Use minimize function to find the min
train = optimizer.minimize(cost_function)
```

# Testing the model

```python
#Initialze all variable
model = tf.global_variables_initializer()
```

```python
with tf.Session() as sess:
    sess.run(model)
    for step in range(0,101):
        sess.run(train)
        if (step%20) ==0:
            plt.plot(x_point,y_point,'o',label='step = {}'.format(step))
            plt.plot(x_point,sess.run(A)*x_point+sess.run(b))
            plt.legend()
            plt.show()
    print('Parameter A =',sess.run(A))
    print('Parameter b = {}'.format(sess.run(b)))
```
